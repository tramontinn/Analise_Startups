```python
Base_Dados['Data de Ades達o'] = pd.to_datetime( Base_Dados['Data de Ades達o'] )

Base_Dados['Data de Ades達o'].head()
```

```python
Base_Dados['Mes'] = pd.DatetimeIndex( Base_Dados['Data de Ades達o'] ).month
Base_Dados['Ano'] = pd.DatetimeIndex( Base_Dados['Data de Ades達o'] ).year

Base_Dados.head()
```

```python
Analise_Agrupada = Base_Dados.groupby( by=['Pais', 'Ano', 'Mes', 'Empresa'] ).count()['Id'].reset_index()

Analise_Agrupada
```

```python
Analise_Agrupada.loc[
  Analise_Agrupada['Pais'] == 'Brazil'
]
```

```python
Base_Dados['Valor ($)'] = pd.to_numeric( Base_Dados['Valor ($)'].apply( lambda Linha: Linha.replace('$', '') ) )

Base_Dados.head()
```

```python
Base_Dados['Data de Ades達o'] = pd.to_numeric( Base_Dados['Data de Ades達o'] )
```

```python
Analise_Pais = Base_Dados.groupby( by=['Pais'] ).sum()['Valor ($)'].reset_index()

Analise_Pais.head()
```

```python
Analise_Valor = Analise_Pais.sort_values('Valor ($)', ascending=False)
Analise_Valor.head()
```

```python
plt.figure( figsize=(15,6) )
plt.plot( Analise_Valor['Pais'], Analise_Valor['Valor ($)'] )
plt.title('Analise do Valor por Pais')
plt.xticks( rotation=45, ha='right');
```
